## Premise: As an employee of Oath, a Verizon company, I have the opportunity to include VZ stock in my 401K portfolio. In determining whether or not to take advantage of this benefit, I am asking the following questions.

1. Where has VZ stock traded historically compared to where it trades now? Could it be over or undervalued?
2. What monthly and yearly trade patterns does Verizon's stock generally follow? Is it typically pretty consistent?
3. What does the last year and last quarter look like for Verizon- is price up? Is volume up too indicating that the price adjustment is true?
4. What did last Q1 look like? Can I expect to make money in this Q1?
5. What daily and monthly fluctations in open, close, high, low, and volume are normal? Is Verizon riskier than other stocks I might invest in?
6. What is the typical range of closing prices for Verizon stock and how does it compare to other Technology and Telecommunications companies?
7. What is the pattern of daily deltas- meaning, how much does VZ stock typically rise or fall within a day as a dollar amount and as a percentage of the opening price? Is this stock volatile at particular times of the year or days of the week? How does that compare to others in the Technology and Telecommunications industry?
8. How does the performance of the companies from the same industy (Tech Telco) impact Verizon?
9. What about companies outside of the industry?

## In order to address the above concerns, I will look at historical stock data from 2006 thru 2017 for Verizon, a few Tech Telco companies, and several other large stocks

```python
import numpy as np
import pandas as pd
```

```python
# read stock market data into df
df = pd.read_csv('all_stocks_2006-01-01_to_2018-01-01.csv')
# exploring the data: preview top 5 rows
df.head()
```

```python
#exploring the data: which companies/stocks are included?
df['Name'].unique()
```

To represent the difference between a stock's Open and Close values within a day, I'll establish the metric "Delta" for change. Since Delta is a dollar amount and stocks have very different dollar valuations, I've also created "Delta%" so that the relative change in a stock's price can be compared to other stocks.

```python
df['Delta'] = df['Close'] - df['Open']
df['Delta%'] = ((df['Close'] - df['Open'])/df['Open'])*100
```

```python
df.head()
```

Next I'll do some data cleaning on Date so that I can create more robust vizualizations of time series data.

```python
import datetime as dt
```

```python
# convert Date from string to datetime
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# add columns for year, month, day
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day

# Column for weekday: Monday is 0 and Friday is 4
df['Weekday'] = df['Date'].dt.weekday

df.head()
```

```python
import calendar

df['monthName'] = df['Month'].apply(lambda x: calendar.month_abbr[x])
df['dayName'] = df['Weekday'].apply(lambda x: calendar.day_abbr[x])

df.head()
```

As a final data cleaning step, I'll set-up a couple of slices of df that will be used in multiple vizualizations.

```python
#establish df for Verizon (VZ)
VZ = df[df['Name'] == "VZ"]
```

```python
#establish df with Tech and Telco companies for comparison against Verizon (VZ)
techTelco = df[(df['Name'] == 'AAPL') | (df['Name'] == 'CSCO') | (df['Name'] == 'IBM') | (df['Name'] == 'INTC') | (df['Name'] == 'MSFT') | (df['Name'] == 'GOOGL') | (df['Name'] == 'AMZN')]
```
